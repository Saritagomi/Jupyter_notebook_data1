```python
import pandas as pd
```

```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
type(data)
```

```python
df = pd.DataFrame(data.data , columns = data.feature_names)
```

```python
df.head()
```

```python
df.head(-1)
```

```python
df['Species'] = data.target
```

```python
df.head()
```

```python
data.target
```

```python
target = np.unique(data.target)
```

```python
target
```

```python
data.target_names
target_names = np.unique(data.target_names)
target_names
```

```python
targets = dict(zip(target,target_names))
targets
```

```python
df['Species'] = df['Species'].replace(targets)
df
```

```python
# Now we will start the process for making the decision tree
```

```python
x = df.drop(columns = 'Species')
y = df['Species']
x
```

```python
x.columns
```

```python
y
```

```python
feature_names = x.columns
feature_names
```

```python
labels = y.unique()
labels
```

```python
from sklearn.model_selection import train_test_split
x_train,test_x,y_train,test_lab=train_test_split(x,y,test_size=0.4,random_state=43)
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(max_depth=3,random_state=42)
clf.fit(x_train,y_train)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(30,10))
a=tree.plot_tree(clf,feature_names=list(feature_names),class_names=list(labels))
```
