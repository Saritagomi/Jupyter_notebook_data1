```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
X = np.array([ 2, 4, 6, 8]).reshape(-1, 1)  # Input features (reshaped to a 2D array when we have one dimension array)
y = np.array([ 3, 7, 5, 10])               # Target values

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)
X_test = np.array([6, 7,10,15]).reshape(-1, 1)  # New data points for prediction
predictions = model.predict(X_test)
print(predictions)
```

```python
# Plot the data and the regression line
plt.scatter(X, y, color='blue', label='Data')
```

```python
plt.plot(X_test, predictions, color='cyan', label='Regression Line')
plt.xlabel('Input')
plt.ylabel('Target')
plt.legend()
plt.show()
```

```python
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X_test, predictions, color='cyan', label='Regression Line')
plt.xlabel('Input')
plt.ylabel('Target')
plt.legend()
plt.show()
```
